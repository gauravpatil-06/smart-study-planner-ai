```python
# ---------------------------------------------------------
# Smart Study Planner – AI Concierge Agent (NO API KEY NEEDED)
# ERROR-FREE KAGGLE VERSION
# ---------------------------------------------------------

import json

# ---------------------------------------------------------
# Custom Tool Simulation (study planner)
# ---------------------------------------------------------

def create_study_plan(subjects: str, hours: int):
    """Generates a simple, evenly-divided study plan."""
    subjects_list = [s.strip() for s in subjects.split(",")]
    
    # Handle division by zero if subject list is empty
    if not subjects_list:
        return "Error: No subjects provided."
        
    per_subject = round(hours / len(subjects_list), 2)

    plan = "📚 Daily Study Plan\n"
    plan += "----------------------\n"
    for s in subjects_list:
        plan += f"- {s}: {per_subject} hours\n"
    return plan

# "Tool" wrapper
def study_planner_tool(subjects, hours):
    """Simulates a tool call, returning a structured dictionary."""
    return {
        "tool_used": "study_planner",
        "subjects": subjects,
        "hours": hours,
        "output": create_study_plan(subjects, hours)
    }

# ---------------------------------------------------------
# Memory Simulation
# ---------------------------------------------------------

memory = {
    "user_name": "Gaurav",
    "preferred_subjects": ["Maths", "Science", "English"],
    "available_hours": 3
}

# ---------------------------------------------------------
# Simulated Agent Response (NO API CALL)
# ---------------------------------------------------------

def agent_simulation():
    """Simulates the agent's logic to fetch memory and call a tool."""
    subjects = ", ".join(memory["preferred_subjects"])
    hours = memory["available_hours"]

    # Agent calls the "tool"
    tool_result = study_planner_tool(subjects, hours)

    # Agent formats the final response
    response = {
        "agent_message": "Here is your personalized study plan based on stored preferences:",
        "memory_used": memory,
        "tool_call": tool_result
    }

    return response

# ---------------------------------------------------------
# OUTPUT
# ---------------------------------------------------------

result = agent_simulation()

print("🤖 AGENT RESPONSE:\n")
print(result["agent_message"])
print("\n🧠 MEMORY USED:")
print(json.dumps(result["memory_used"], indent=2))
print("\n🛠 TOOL EXECUTION RESULT:")
print(result["tool_call"]["output"])
```
